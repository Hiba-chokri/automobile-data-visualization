```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(color_codes=True)
%matplotlib inline
```

```python
auto = pd.read_csv(r'C:\Users\cap\Downloads\clean_auto_data.csv')
```

```python
auto.head()
```

```python
#disturbution of one variable
sns.displot(auto['normalized-losses'])
```

```python
sns.displot(auto['city-mpg'],kde=False,rug=True)
```

```python
#bivariable
#diff kinds(scatter,resid,hex,rug,kde)
sns.jointplot(auto['wheel-base'],auto['length'], kind="hex")
```

```python
sns.pairplot(auto[['normalized-losses','wheel-base','length']])
```

```python
#a categorecal,and continues variables
sns.stripplot(auto['horsepower-binned'],auto['price'])
```

```python
sns.boxplot(auto['horsepower-binned'],auto['wheel-base'], hue= auto['num-of-doors'])
```

```python
sns.barplot(auto['body-style'], auto['highway-L/100km'],hue=auto['engine-location'])
```

```python
#factorplot
#vration types of kind input(bar,strip,point,count,violin)
sns.catplot(x="horsepower-binned",
               y="price",
               col="engine-location",
               data=auto,
               kind="swarm")
```

```python
sns.lmplot(x="city-L/100km",y="normalized-losses",data=auto,hue="drive-wheels")
```
